```python
import os
import json
import numpy as np
import pandas as pd
import altair as alt
```

```python
path_root_data = os.path.join('..', 'Data')
path_vis_data = os.path.join(path_root_data, 'Visualization_data', 'data', 'multiple_datasets',
                             'vis', 'mds_1_Overview')
path_vega_lite_spec = os.path.join(path_vis_data, 'mds_1_Overview.json')
path_source_data = os.path.join(path_vis_data, 'hm_source_data.json')
```

```python
with open(path_vega_lite_spec, 'r') as tmp_file:
    dict_vega_lite_spec = json.load(tmp_file)

chart_result = alt.Chart.from_dict(dict_vega_lite_spec)
df_source_data = pd.read_json(path_source_data)
chart_result.data = df_source_data
chart_result
```

```python
df_source_data
```

```python
df_source_data['Encoding_max'] = df_source_data[
    'Encoding_max'].str.replace('_levels_1_and_2_encoding','')
df_source_data['Encoding_max'] = df_source_data[
    'Encoding_max'].str.replace('seq','CMANGOES_seq_')
df_source_data['Encoding_max'] = df_source_data[
    'Encoding_max'].str.replace('str','CMANGOES_str_')

df_source_data
```

```python
chart_result = alt.Chart(
    df_source_data, title="CMANGOES multi-dataset results"
).mark_rect().encode(
    x=alt.X('Encoding_max', type='nominal', axis=alt.Axis(labelAngle=-45)),
    y=alt.Y('Dataset', type='nominal'),
    color=alt.Color('F1', type='quantitative', scale=alt.Scale(scheme='blues')),
    tooltip=['Dataset', 'Encoding_max', 'F1', 'is_imbalanced', 'bio_field']
)

chart_result
```
